```python
import requests
import xml.etree.ElementTree as ET
from urllib.request import urlopen
```

```python
#here we are using url page to take data and show its texts
url = "https://www.tcmb.gov.tr/kurlar/today.xml"
resp = requests.get(url=url)
print(resp.text)
```

```python
tree =ET.parse(urlopen(url))
root = tree.getroot()
```

```python
#firstly adding all currency to liste then divde them  show selling and buying banknote

liste =[]
liste.append(root.findall('Currency'))
for i in liste[0]:
    currencyCode = i.get('Kod')
    banknotBuying = i.find("BanknoteBuying").text
    banknotSelling = i.find("BanknoteSelling").text
    ForexBuying = i.find("ForexBuying").text
    ForexSelling = i.find("ForexSelling").text
    name = i.find("Isim").text
    if currencyCode == "USD":
        result =float(banknotSelling) - float(banknotBuying)
        print("USD",banknotSelling)
        print("USD",banknotBuying)
        print(" Banka alıs satıs arasındakı kur farkı ->",str(result))
    if currencyCode == "EUR":
        result =float(banknotSelling) - float(banknotBuying)
        print("EUR",banknotSelling)
        print("EUR",banknotBuying)
        print(" Banka alıs satıs arasındakı kur farkı ->",str(result)) 
    if currencyCode == "QAR":
        result =float(ForexSelling) - float(ForexBuying)
        print("QAR",ForexBuying)
        print("QAR",ForexSelling)
        print("Para birimi adı : ",name) 
        print(" Banka alıs satıs arasındakı kur farkı ->",str(result))        
        
```
