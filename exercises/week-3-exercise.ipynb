```python

plt.figure(figsize=(12, 8))
plt.plot(finance['Year'], finance['Less_than_$5000'], label='Less than $5000')
plt.plot(finance['Year'], finance['$5000_to_$9999'], label='$5000 to $9999')
# Add similar lines for other income ranges
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.title('Income Distribution Over the Years')
plt.legend()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(x='cd', y='Less_than_$5000', data=finance, label='Less than $5000')
sns.barplot(x='cd', y='$5000_to_$9999', data=finance, label='$5000 to $9999')
finance.sort_values(by='cd', inplace=True)

plt.xlabel('Category')
plt.ylabel('Number of People')
plt.title('Income Distribution by Category')
plt.legend()
plt.show()
```

```python
finance['Total_Income'] = finance.iloc[:, 2:].sum(axis=1)
```

```python
grouped_finance=finance.groupby('cd').sum().reset_index()
```

```python
grouped_finance.head(10)
```

```python
finance.columns
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='cd', y='Less_than_$5000', data=grouped_finance, label='Less than $5000')
sns.barplot(x='cd', y='$5000_to_$9999', data=grouped_finance, label='$5000 to $9999')
plt.xlabel('Category')
plt.ylabel('Number of People')
plt.title('Income Distribution by Category (Grouped)')
plt.legend()
plt.show()
```

```python
grouped_finance.info()
```

```python
# Group data by income range
income_groups = grouped_finance.groupby(["cd", "Year"]).sum()["Total_Income"]

# Create bar graph
plt.figure(figsize=(10, 6))
income_groups.unstack().plot(kind="bar", stacked=True)

# Add labels and title
plt.xlabel("County")
plt.ylabel("Total Income")
plt.title("Total Income by County and Year")

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation between the chosen columns
correlation = finance['Less_than_$5000'].corr(finance['Total_Income'])

# Plotting the scatter plot for the two chosen columns
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Less_than_$5000', y='Total_Income', data=finance)
plt.title(f'Scatter Plot between Less_than_$5000 and Total_Income')
plt.show()

# Print the correlation coefficient
print(f"Correlation between Less_than_$5000 and Total_Income: {correlation}")
```

```python
selected_columns = [
    'Total_Agriculture_forestry_fishing_hunting_mining',
    'Total_Construction',
    'Total_Manufacturing',
    'Total_Finance_insurance_realestate_rental_leasing',
]

# Subset DataFrame
subset_df = industry[['Year'] + selected_columns]

# 1. Line Plot
plt.figure(figsize=(12, 6))
for column in selected_columns:
    plt.plot(subset_df['Year'], subset_df[column], label=column)

plt.title('Trends in Selected Sectors Over Time')
plt.xlabel('Year')
plt.ylabel('Total')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Choose a subset of years for better readability
subset_years = [2020]

# Filter DataFrame based on the selected years
subset_df = subset_df[subset_df['Year'].isin(subset_years)]

# Bar Plot
subset_df.plot(x='Year', y=selected_columns, kind='bar', stacked=True, figsize=(12, 6))
plt.title('Total in Selected Sectors Over Time')
plt.xlabel('Year')

plt.ylabel('Total')
plt.show()
```

```python
correlation_matrix = subset_df[selected_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
pairplot_columns = ['Total_Agriculture_forestry_fishing_hunting_mining', 'Total_Finance_insurance_realestate_rental_leasing']
sns.pairplot(subset_df[pairplot_columns])
plt.suptitle('Pairplot for Selected Sectors', y=1.02)
plt.show()
```
