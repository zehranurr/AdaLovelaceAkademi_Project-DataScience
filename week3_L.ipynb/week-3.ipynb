### Data Visualization

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
df = pd.DataFrame({
    'Group' : ['A','B','C','D','E'],
    'Value': [1,5,4,3,9]
    
    
    
})
```

```python
plt.barh(y=df.Group, width = df.Value)
```

```python
df = df.sort_values(by =['Value'])
my_plot = plt.barh(
            y=df.Group,
            width = df.Value,
            color = mcolors.TABLEAU_COLORS,
            edgecolor ='black',
            linewidth= 3,
    )
plt.bar_label(container=my_plot,padding=4)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python

sns.barplot(
    x = tips.total_bill,
    y = tips.day,
    data = tips,
    color="cyan"  
)
plt.figure(figsize=(18,8))
```

```python
g = sns.FacetGrid(tips,col="day",height=7,aspect=5)
g.map(sns.barplot,"sex","total_bill",order=["Male","Female"])
```

```python
total = tips.groupby('day')['total_bill'].sum().reset_index()
```

```python
total
```

```python
barl =sns.barplot(x = 'day', y = 'total_bill',data=total,color='darkblue')
```

```python
smoker = tips[tips.smoker =='Yes']
```

```python
bar2 = sns.barplot(
    x = 'day',
    y = 'total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue'
    
    
)
```

```python
barl =sns.barplot(x = 'day', y = 'total_bill',data=total,color='darkblue')
bar2 = sns.barplot(
    x = 'day',
    y = 'total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue'
    
    
)
top_bar = mpatches.Patch(color='darkblue',label='non-smoker')
bottom_bar= mpatches.Patch(color='lightblue',label='smoker = Yes')
plt.legend(handles=[top_bar,bottom_bar])
plt.show()
```
