```python
df_infected.drop(columns=['Lat','Long','Province/State'],inplace=True)
```

```python
df_infected.head()
```

```python
df_deaths=df_deaths.drop(columns=['Lat','Long','Province/State'])
df_recovered=df_recovered.drop(columns=['Lat','Long','Province/State'])
```

```python
df_deaths.head()
```

```python
def make_frame(country:str):
    """ Take country name as parameter,return infected,recovered,deaths values as timeseries
        pd.DataFrame()
    """
    
    df = pd.DataFrame({
        'infected': df_infected.loc[country],
        'recorved' : df_recovered.loc[country],
        'deaths' : df_deaths.loc[country]
    })
    df.index = pd.to_datetime(df.index , format = '%m/%d/%y')
    return df
```

```python
df_us =make_frame('US')
```
